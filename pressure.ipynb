#  Previsão da perda de carga na saída de um riser a partir da velocidade do ar na entrada e da vazão mássica de sólidos

Este modelo permite estimar a perda de carga no ciclone de primeiro estágio de uma unidade piloto de leito fluidizado circulante gás-sólido para diferentes velocidades de entrada. No riser circulam ar ambiente e partículas de FCC pertencentes à classe B de Geldart. O valor de perda de carga encontrado é usado para definir a condição de contorno de saída em simulações de CFD (Computational Fluid Dynamics). Assim evita-se usar a condição de pressão atmosférica na saída, que induz a um erro porque a saída não se dá para a atmosfera, mas para o ciclone de primeiro estágio. O dataset utilizado contém medições realizadas na própria unidade piloto e os dados foram coletados usando o software Labview.

### Representação do equipamento estudado

Para que o problema possa ser representado matematicamente e a simulação computacional seja feita, é necessário impor condições de contorno condizentes com o processo real. Conhecendo-se a velocidade do ar na entrada do riser, que pode ser medida através de um medidor do tipo Venturi, pode-se determinar o fluxo mássico de sólidos a partir de uma curva de calibração da vazão. Já a perda de carga na saída é determinada através de medições realizadas na saída do riser para o ciclone com um tubo de Pitot. Como a geometria do equipamento é complexa, torna-se mais difícil calcular a perda de carga para diferentes velocidades. Para evitar a necessidade de se determinar experimentalmente a perda de carga para cada velocidade desejada, construiu-se uma curva de calibração com dados coletados na unidade piloto.

## Treinamento e avaliação de modelos de regressão

```python
# Divisão do conjunto de dados entre variáveis preditoras (x) e variável predita (y)
x = df.drop('Delta_P1 [Pa]', axis=1)
y = df['Delta_P1 [Pa]']
```

```python
# Divisão do conjunto de dados em treino e teste, com 20% dos dados para teste
from sklearn.model_selection import train_test_split
np.random.seed(42)

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
# Verificação do formato dos conjuntos de teste
x_test.shape, y_test.shape
```

```python
# Importação dos algoritmos para os modelos Random Forest e Regressão Linear
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
# Treinamento dos modelos e avaliação das métricas
models = {"Regressão Linear": LinearRegression(),
          "Random Forest": RandomForestRegressor()}

def fit_and_score(x_train, y_train, x_test, y_test):
  
  np.random.seed(42)
  
  model_scores = {}

  for name, model in models.items():
    model.fit(x_train, y_train)

    model_scores[name] = model.score(x_test, y_test)

  return model_scores
```

```python
model_scores = fit_and_score(x_train, y_train, x_test, y_test)
model_scores
```

```python
# Comparação da performance dos modelos
model_compare = pd.DataFrame(model_scores, index= ['Accuracy'])
fig, ax = plt.subplots(figsize=(8,6))
model_compare.plot.barh(ax=ax)
plt.title('Score comparison between the models')
plt.xticks(rotation=0);
```

Os dois modelos têm performance semelhante, com ajuste de 98,4% para a Regressão Linear e 98,9% para o Random Forest. O modelo Random Forest será usado nos testes seguintes.

## Tunagem de hiperparâmetros

A otimização dos hiperparâmetros do modelo será feita usando os algoritmos RandomizedSearchCV e GridSearchCV.

```python
# Importação das bibliotecas
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
# Tunagem de hiperparâmetros
rf_grid = {"n_estimators": np.arange(10,1000,50),
           "max_depth": [None, 3, 5, 7, 10],
           "min_samples_split": np.arange(2,20,2),
           "min_samples_leaf": np.arange(1, 25, 1)}
```

```python
# Teste de diferentes cenários com RandomizedSearchCV
rs_rf = RandomizedSearchCV(RandomForestRegressor(),
                           param_distributions = rf_grid,
                           cv=5,
                           n_iter = 20,
                           verbose=True
                           )
rs_rf.fit(x_train, y_train)
```

```python
# Melhores parâmetros encontrados
rs_rf.best_params_
```

```python
# Com os melhores parâmetros encontrados usando o RandomizedSearchCV, faz-se uma nova otimização usando o GridSearchCV
rs_rf_grid = {"max_depth": [7],
              "min_samples_leaf": [1,2,3],
              "min_samples_split": [3,4,5],
              "n_estimators": [510]}
gs_rf = GridSearchCV(RandomForestRegressor(),
                     param_grid = rs_rf_grid,
                     cv= 5, 
                     verbose=True)

gs_rf.fit(x_train, y_train)
```

```python
# Melhores parâmetros do GridSearchCV
gs_rf.best_params_
```

```python
# Performance do modelo otimizado
gs_rf.score(x_test, y_test)
```

```python
# Avaliação do modelo
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
y_preds = gs_rf.predict(x_test)
```

```python
tabela_auxiliar = pd.DataFrame()
tabela_auxiliar["y_teste"] = y_test
tabela_auxiliar["Random Forest"] = y_preds

plt.figure(figsize=(15,5))
sns.lineplot(data=tabela_auxiliar)
plt.show()
```

 O gráfico acima compara os valores do conjunto de perda de carga do conjunto de teste do dataset com os valores previstos usando o modelo Random Forest otimizado. Apesar de haver algumas diferenças nos picos, o modelo é suficiente para fazer estimativas da perda de carga no topo do riser em diferentes condições de velocidade e fluxo mássico na entrada.

```python
y_preds = gs_rf.predict([[10,46]])
y_preds
```

O resultado acima é uma previsão da perda de carga, em Pascal, na saída do riser para uma velocidade de entrada de 10 m/s e um fluxo mássico de sólidos de 46 kg/m²s. O valor obtido de 160,7 Pa pode ser usado como condição de contorno na saída do riser em simulações de CFD.
